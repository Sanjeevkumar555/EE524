```python
# question 1 (a) #
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 100 samples of x
x = np.linspace(-10,10,100)

#parameter for Normal Distribution
mu, sigma = 0, 1

# Plot the distribution curve
plt.plot(x,1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (x - mu)**2 / (2 * sigma**2) ))

plt.xlabel('x')
plt.ylabel("f(x)")
plt.title("probability density function with paramerter N(0,1)")
plt.show()


#parameter for Normal Distribution
mu, sigma = 3, 2

# Plot the distribution curve
plt.plot(x,1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (x - mu)**2 / (2 * sigma**2) ))

plt.xlabel('x')
plt.ylabel("f(x)")
plt.title("probability density function with paramerter N(3,2)")
plt.show()
```

Here, for different cases we have different mean and different standard deviations, for less value of standard 
deviation pdf curve will be more sharp but for high value of standard deviation curve will be flatter. but still
in both the case integration of pdf from *minus infinity* to *plus infinity* will be **one**

```python
# question 1 (b) #
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# limits of x1 and x2
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)

# parameter for two dimensional nprmal random variable
mu1, mu2 = 0, 0
sigma1, sigma2 = np.sqrt(1), np.sqrt(2)
rho = 0.6/np.sqrt(2)

# pdf of two dimensional normal random variable
z = (1/2*np.pi*sigma1*sigma2*np.sqrt(1-rho**2))*np.exp((-1/2*(1-rho**2))*((x1-mu1)**2/sigma1**2 + \
          (x2-mu2)**2/sigma2**2 - 2*rho*(x1-mu1)*(x2-mu2)/sigma1*sigma2))

# plot of pdf
x1,x2=np.meshgrid(x1,x2)
fig = plt.figure() 
ax = fig.gca(projection='3d')
ax.plot_surface(x1, x2, z, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("z")

plt.title('two dimensional normal random variable ')

cset = ax.contourf(x1, x2, z, zdir='z', offset=-0.50, cmap=cm.viridis)
plt.show()
```
